```python
def doAvgSave(chunk):
    df1 = pd.read_csv(chunk[0])
    tmp = df1.copy()
    tmp = tmp.drop(columns=['cc1'] )
    #print tmp
    for i in range(1,len(chunk)):
        tmp1 = pd.read_csv(chunk[i])
        tmp1 = tmp1.drop(columns=['cc1'] )
        tmp = tmp + tmp1 # add those     
    tmp = tmp.div( len(chunk)) # average
#     plt.plot(tmp.iloc[8,1:].values)
#     plt.grid(True)
#     plt.show()
    df1.update(tmp) # this is the new data frame with average 
    # write to some location 
    plt.plot(df1.iloc[8,1:].values*1e9)
    plt.grid(True)
    plt.show()
```
