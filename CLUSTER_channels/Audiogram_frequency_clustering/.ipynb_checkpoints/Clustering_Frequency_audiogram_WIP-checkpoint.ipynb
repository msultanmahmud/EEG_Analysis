# Construct scatter matrices (within-class scatter matrix and bet-class scatter matrix) for LDA

```python
# ldadata.values
```

```python
from sklearn.preprocessing import LabelEncoder
X=ldadata=x.iloc[2:34,10:22].values
y=x.iloc[2:34,-1].values
enc = LabelEncoder()
label_encoder = enc.fit(y)
y = label_encoder.transform(y) + 1
label_dict = {1: 'NH', 2: 'HI'}
```

```python
Xf=X.astype(float)
Xf
y
```

```python
np.set_printoptions(precision=4)
mean_vectors = []
for cl in range(1,3):
    mean_vectors.append(np.mean(Xf[y==cl], axis=0))
    print 'Mean Vector class %s: %s\n' %(cl, mean_vectors[cl-1])
```

```python
# NH=(mean_vectors[0][0:6]+mean_vectors[0][6:12])/2.0
# HI=(mean_vectors[1][0:6]+mean_vectors[1][6:12])/2.0
# plt.plot(-NH, c='red')
# plt.plot(-HI, c='b')
# plt.grid(True)
# plt.ylim([-90, 0])
# plt.show()
```

```python
d=12 # number of features
S_W = np.zeros((d,d))
for cl,mv in zip(range(1,3), mean_vectors):
    class_sc_mat = np.zeros((d,d))                  # scatter matrix for every class
    for row in Xf[y == cl]:
        row, mv = row.reshape(d,1), mv.reshape(d,1) # make column vectors
        #print (row-mv).dot((row-mv).T)
        class_sc_mat += (row-mv).dot((row-mv).T)
    S_W += class_sc_mat                             # sum class scatter matrices
print'within-class Scatter Matrix:\n', S_W
```

```python
# d=12 # number of features
# S_W = np.zeros((d,d))
# for cl,mv in zip(range(1,3), mean_vectors):
#     class_sc_mat = np.cov(Xf[y==cl].T)                  # scatter matrix for every class
#     S_W += class_sc_mat                             # sum class scatter matrices
# print'within-class Scatter Matrix:\n', S_W
```

```python
d=12;
overall_mean = np.mean(Xf, axis=0)
S_B = np.zeros((12,12))
for i,mean_vec in enumerate(mean_vectors):  
    n = Xf[y==i+1,:].shape[0]
    mean_vec = mean_vec.reshape(d,1) # make column vector
    overall_mean = overall_mean.reshape(d,1) # make column vector
    S_B += n * (mean_vec - overall_mean).dot((mean_vec - overall_mean).T)

print'between-class Scatter Matrix:\n', S_B
```

```python
eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
for i in range(len(eig_vals)):
    eigvec_sc = eig_vecs[:,i].reshape(12,1)   
    print('\nEigenvector {}: \n{}'.format(i+1, eigvec_sc.real))
    print('Eigenvalue {:}: {:.2e}'.format(i+1, eig_vals[i].real))
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)
# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in decreasing order:\n')
for i in eig_pairs:
    print(i[0])
```

```python
# print('Variance explained:\n')
# eigv_sum = sum(eig_vals)
# for i,j in enumerate(eig_pairs):
#     print('eigenvalue {0:}: {1:.2%}'.format(i+1, (j[0]/eigv_sum).real))
```

```python
d=12;
W = np.hstack((eig_pairs[0][1].reshape(d,1), eig_pairs[1][1].reshape(d,1),eig_pairs[2][1].reshape(d,1),eig_pairs[3][1].reshape(d,1)))
print('Matrix W:\n', W.real)
```
