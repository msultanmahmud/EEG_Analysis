```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
path='/home/sultan/EEG/CLUSTER_channels/Audiogram_frequency_clustering/data_audiogram/'
x = pd.read_excel(path+'class_labelRotmanSIN_subjects_demo_behav.xls')
aud=x.iloc[2:34,10:22]
aud
data=aud
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
#iris = load_iris()
#X = pd.DataFrame(iris.data, columns=iris['feature_names'])
#print(X)
#data = X[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']]
sse = {}
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.title("Aging dataset")
#plt.grid(True)
plt.show()
```

# K means on IRISH DATA

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris['feature_names'])
#print(X)
data1 = X[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']]
sse = {}
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data1)
    data1["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.title("Iris dataset")
plt.show()
```

# PCA analysis

```python
from sklearn.decomposition import PCA
import seaborn as sns; sns.set()
```

```python
pca = PCA(4)  # project from 64 to 2 dimensions
projected = pca.fit_transform(ldadata)
print(ldadata.shape)
print(projected.shape)
```
