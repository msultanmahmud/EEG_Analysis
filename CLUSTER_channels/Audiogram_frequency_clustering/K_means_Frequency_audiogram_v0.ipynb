# K-means Clustering

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 41)
y_kmeans = kmeans.fit_predict(data.values)
y_kmeans
```

```python
# y_kmeans=np.where(y_kmeans==0,1,0);
df1 = pd.DataFrame({'True_label':ylabel,'predic_label':y_kmeans})
# df1
# this gives 1 where val is 0 and gives 0 where val is anything else, in your case 1
# np.where(y_kmeans==0,1,0);
```

```python
print (classification_report(y_kmeans, ylbl.values))
```

```python
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_kmeans, ylbl.values)
roc_acu=roc_auc = auc(false_positive_rate, true_positive_rate)
print round(roc_acu,2)
# print("ROC_AUC: {} is {}".format(a[cn], round(roc_acu,2)))
```

```python
y_test,pred=ylbl.values,y_kmeans
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,pred)
#false_positive_rate, true_positive_rate, thresholds = metrics.roc_curve(pred, y_test)
#print false_positive_rate
roc_auc = auc(false_positive_rate, true_positive_rate)
#print 'FPR=',false_positive_rate,'\n','TPR=',true_positive_rate
fig,ax1=plt.subplots(ncols=1)
ax1.plot(false_positive_rate, true_positive_rate, 'b-', label = 'AUC = %0.2f' % roc_auc)
ax1.legend(loc = 'lower right')
ax1.plot([0, 1], [0, 1],'r--')
ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1])
ax1.set_ylabel('True Positive Rate')
ax1.set_xlabel('False Positive Rate')
ax1.set_title('Receiver Operating Characteristic')
```
